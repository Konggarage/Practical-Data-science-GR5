```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap

from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    classification_report, roc_curve
)
```

```python
train_df = pd.read_csv("/Users/kongphopkaochot/Desktop/Senior_Project/Practical Datascience/week15/TRAIN_DATASET.csv")
test_df = pd.read_csv("/Users/kongphopkaochot/Desktop/Senior_Project/Practical Datascience/week15/TEST_DATASET.csv")
```

```python
X_train = train_df.drop(columns=["appendicitis", "Migration"])
y_train = train_df["appendicitis"]

X_test = test_df.drop(columns=["appendicitis", "Migration"])
y_test = test_df["appendicitis"]
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

acc_scores = []
f1_scores = []

for train_idx, val_idx in kf.split(X_train):
    X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    model = LogisticRegression(max_iter=1000)
    model.fit(X_tr, y_tr)

    y_pred = model.predict(X_val)

    acc_scores.append(accuracy_score(y_val, y_pred))
    f1_scores.append(f1_score(y_val, y_pred))

print("K-Fold Accuracy:", np.mean(acc_scores))
print("K-Fold F1:", np.mean(f1_scores))
```

```python
final_model_no_Migration = LogisticRegression(max_iter=1000)
final_model_no_Migration.fit(X_train, y_train)
```

```python
y_pred_test = final_model_no_Migration.predict(X_test)
y_prob_test = final_model_no_Migration.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred_test))
print("AUC:", roc_auc_score(y_test, y_prob_test))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_test))
```

```python
intercept = final_model_no_Migration.intercept_[0]
coefficients = final_model_no_Migration.coef_[0]
features = X_train.columns

print("\n=== Logistic Regression Result ไม่มี TEMP ===")
print(f"Intercept: {intercept:.4f}")
print("-" * 45)

for name, coef in zip(features, coefficients):
    print(f"Weight of {name:15}: {coef: .4f}")

print("-" * 45)

equation = f"Logit (z) = {intercept:.4f}"

for name, coef in zip(features, coefficients):
    sign = "+" if coef >= 0 else "-"
    equation += f" {sign} ({abs(coef):.4f} * {name})"

print(equation)
```

```python
for col in X_train.columns:
    corr = train_df[[col, "appendicitis"]].corr().iloc[0, 1]
    print(col, corr)


explainer = shap.Explainer(final_model_no_Migration, X_train)
shap_values = explainer(X_test)

shap.summary_plot(shap_values, X_test)
```

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# MODEL no Migration ใช้ X_test
y_prob_model_no_temp = final_model_no_Migration.predict_proba(X_test)[:, 1]

# DOCTOR equation no Migration
Baseline_z_no_temp = (
    -3.37
    + 1.04 * X_test["Progression"]
    # + 0.80 * X_test["Migration"]
    + 1.64 * X_test["TEMP"]
    + 0.78 * X_test["Aggravation"]
    + 1.53 * X_test["Rebound"]
    + 0.91 * X_test["WBC"]
    + 0.69 * X_test["Neut"]
)

Baseline_prob_no_temp = sigmoid(Baseline_z_no_temp)

# ROC + AUC
fpr_model, tpr_model, _ = roc_curve(y_test, y_prob_model_no_temp)
auc_model = roc_auc_score(y_test, y_prob_model_no_temp)

fpr_doc, tpr_doc, _ = roc_curve(y_test, Baseline_prob_no_temp)
auc_doc = roc_auc_score(y_test, Baseline_prob_no_temp)

plt.figure(figsize=(6, 6))
plt.plot(fpr_model, tpr_model, label=f"Model No Migration (AUC={auc_model:.2f})")
plt.plot(fpr_doc, tpr_doc, label=f"Baseline No Migration (AUC={auc_doc:.2f})")
plt.plot([0, 1], [0, 1], "--")

plt.title("ROC Comparison - NO Migration")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
```
